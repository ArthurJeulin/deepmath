```python
# Test Matplotlib
# N(0,1)
def f(x):
    return (1/np.sqrt(2*np.pi))*np.exp(-(x**2)/2)

# Interval
a, b = -5, 5
X = np.linspace(a,b,num=100)
Y = f(X)
# Plot only point
#plt.scatter(X,Y)
plt.plot(X,Y)
plt.title('Gaussian')
#plt.axis('equal') # repère orthonormé
plt.grid()
#plt.xlim(a,b) bornes de l'axes x
#plt.save('gaussian.png')
plt.show()
```

```python
# Plot sigma function
def f(x,k):
    return 1/(1+np.exp(-k*x))
# Interval
a,b = -5,5
X = np.linspace(a,b,num=100)
for k in range(1,5):
    Y = f(X,k)
    plt.plot(X,Y, label="k={}".format(k))

plt.legend()
plt.show()
```

```python
# Numpy with 2 variables
delta = 0.2
VX = np.arange(-2.5,2.5,delta)
VY = np.arange(-2.5, 1.7, delta)
# Create a grid
X,Y = np.meshgrid(VX,VY)
print(X)
def f(x,y):
    return y**3 + 2*y**2 - x**2
Z = f(X,Y)

# Graph of f

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z)

ax.view_init(30,-135)
ax.view_init(5, -145)
plt.show()
```

```python
# otherway to do 3D plot
def f(x,y):
    return 1/(2 + x**2 + y**2)
n = 20
a,b = -5, 5
VX = np.linspace(a,b,n)
VY = np.linspace(a,b,n)
X,Y = np.meshgrid(VX,VY)
Z = f(X,Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(40,-70)
ax.plot_wireframe(X,Y,Z)
plt.show()
```

```python
# afficher les lignes de niveaux

def f(x,y):
    return np.sin(x) + np.cos(y)
n = 100
a,b = -8, 8
VX = np.linspace(a,b,n)
VY = np.linspace(a,b,n)
X,Y = np.meshgrid(VX,VY)
Z = f(X,Y)

nb_niveau = 10
plt.contour(X,Y,Z, nb_niveau)
plt.axis('equal')
plt.show()
```
