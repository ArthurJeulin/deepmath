```python
# Includes
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Play with np vector
# Simplest array
X = np.array([1,2,3,4])
print(X)
# Create a vector from 1 and increment of 0.5 until reaching 10
Y = np.arange(1,10,0.5)
print(Y)
# Interval from 0 to 1
Z = np.linspace(0,1,num=5)
print(Z)
print("Sum of X: " + str(np.sum(X)))

U = np.zeros(10)
print(U)
T = np.ones(13)
print(T)

# Create a vector of 5 random numbers between 0 to 1 
R = np.random.random(5)
print(R)
# Print element of R
for r in R:
    print(r)
print("---------------------")
# Other way to iterate over the list
for i in range(len(R)):
    print(i, R[i])
# Applied basic math function over a vector
print(np.square(X))
# Do not use math module if you use numpy function
```

```python
# Test Matplotlib
# N(0,1)
def f(x):
    return (1/np.sqrt(2*np.pi))*np.exp(-(x**2)/2)

# Interval
a, b = -5, 5
X = np.linspace(a,b,num=100)
Y = f(X)
# Plot only point
#plt.scatter(X,Y)
plt.plot(X,Y)
plt.title('Gaussian')
#plt.axis('equal') # repère orthonormé
plt.grid()
#plt.save('gaussian.png')
plt.show()
```

```python
# Easy plot at one variable
X = np.linspace(0,10,num=100)
Y = np.cos(X)
plt.plot(X,Y)
plt.show()
```
